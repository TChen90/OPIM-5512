### Concatenate and order two series

#### Obtain every alternative number

```python
a = pd.Series(range(10,20,2))
a
```

```python
b = pd.Series(range(11,21,2))
b
```

#### Concatenate and sort values

```python
ab = pd.concat([a,b])
ab
```

```python
ab.sort_values(0, inplace=True) # {0 or ‘index’, 1 or ‘columns’}, default 0
                                # Axis to be sorted.
ab
```

### Fix the index values

```python
ab.reset_index(drop=True, inplace=True)
ab
```

### Assign NaN (null values) based on a filter

```python
df = pd.read_csv('F:/UConn/03 Summer 2019/OPIM 5512/Lecture/data/401K.csv')
df.head()
```

```python
df1 = df.copy()
```

```python
df1.loc[(df1.sole == 0), 'prate'] = np.nan
df1.head()
```

```python
df1 = df1.fillna(0)
df1.head()
```

```python
# Assign NaN to mrate and totpart
df1.loc[df.mrate < 1.0, 'mrate'] = np.nan
df1.loc[df.totpart < 1000, 'totpart'] = np.nan
df1.head()
```

### Exercise
Produce a table containing all students and all courses.
For students who did take a course, indicate the grade.
For students who did not take a course, indicate 'NA'.

```python
c1 = c.copy()
s1 = s.copy()
c1['key'] = 1
s1['key'] = 1
cs1 = pd.merge(c1,s1, on='key')
cs1
```

```python
csg = pd.merge(cs1, g, on=['course_id', 'student_id'], how='left')
csg
```

```python
csg.drop('key', axis=1, inplace=True)
csg.head()
```

```python
csg['grade'] = csg['grade'].fillna('NA')
csg.head()
```
